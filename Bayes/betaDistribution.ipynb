```python
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
def getPDF(a, b, prt=False):
    x = np.linspace(0,1,100)
    pdf = beta.pdf(x, a, b)
    if prt:
        plt.plot(x,pdf)
    return pdf
```

```python
def updatePDF(a, b, z, N, prt=False):
    a = z + a
    b = N - z + b
    return getPDF(a,b, prt)
```

```python
# Set the Prior; assume a fair coin
a = 50
b = 50
pdf = getPDF(a, b, True)
```

```python
# What is the PDF and posterior if it comes up heads 80 times in 100 flips?
z = 80
N = 100
pdf = updatePDF(a, b, z, N, True)
```
