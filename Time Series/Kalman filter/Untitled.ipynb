```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df    = sm.datasets.macrodata.load_pandas().data
index = pd.DatetimeIndex(start='1959Q1', end='2009Q4', freq='Q')
df.set_index(index, inplace=True)
```

```python
col = "realgovt"
cycle, trend = sm.tsa.filters.hpfilter(df[col], 1600)

ts = pd.DataFrame(df[col])
ts["cycle"] = cycle
ts["trend"] = trend
```

```python
fig, ax = plt.subplots()
ts[["realgovt", "trend"]]["2000-03-31":].plot(ax=ax,fontsize=16)
plt.show()
```

```python
recent = ts[ts.index > "2008-09-30"]
plt.plot(recent["cycle"])
plt.show()
```
