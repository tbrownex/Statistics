```python
from getConfig import getConfig
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def getPredictions(count):
    fname = "predictions.csv"
    return np.loadtxt(config["dataLoc"]+fname, delimiter=",")
```

```python
def plot(data):
    test = data[:,0]
    pred = data[:,1]
    plt.plot(test, label="Test")
    plt.plot(pred, label="Prediction")
    plt.legend(loc='upper left')
    plt.ylim(0, 1300)
    plt.show()
```
