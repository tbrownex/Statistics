#### Periodogram
Create a time series with 2 seasonal components: weekly and monthly.
Add noise and see if you can identify the seasonalities

```python
# Create a time series with 2 seasonal components: one at frequency 12 (monthly) 
# and the other at frequency 52 (weekly)

import matplotlib.pyplot as plt
import math
from scipy import signal
import numpy as np
```

```python
A1 = .866          # Amplitudes
A2 = .5
F1 = 12            # Monthly
F2 = 52            # Weekly
N  = 10000         # Number of observations
INTERVAL = 10000   # observations/second
```

```python
time = np.arange(N) / INTERVAL
noise = np.random.normal(scale=0.5, size=N)
```

```python
# First component
x = A1*np.sin(2*np.pi*F1*time)
# Second component
x += A2*np.sin(2*np.pi*F2*time)
# Noise
x += noise
```

```python
plt.plot(x)
plt.show()
```

```python
f, Pxx_den = signal.periodogram(x, fs=10000)
plt.semilogy(f[:60], Pxx_den[:60])
plt.ylim([1e-7, 1e2])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
