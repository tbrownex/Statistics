##### We captured both the AICC and the MAPE. Either one can perform best at a given time series so test them both

```python
topCount = 25
lowestError = df.sort_values(by=["error"], ascending=True)[:topCount]
lowestAICC = df.sort_values(by=["aicc"], ascending=True)[:topCount]
```

```python
print("{:>22}{:>15}".format("AICC", "Error"))
for x in ["alpha", "beta", "gamma"]:
    print("{:<15}{:<15.3f}{:.3f}".format(x, lowestAICC[x].mean(), lowestError[x].mean() ))
```
