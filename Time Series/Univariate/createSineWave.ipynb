#### Creates a very simple time series that should be easily forecast-able

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Amplitudes
A1 = .866          
A2 = .5
# Frequencies
F1 = 12            # Monthly
#F2 = 52            # Weekly

N  = 1095         # Number of observations
INTERVAL = 365   # observation frequency

time = np.arange(N) / INTERVAL
noise = np.random.normal(scale=0.5, size=N)

# First component
x = A1*np.sin(2*np.pi*F1*time)
# Second component
#x += A2*np.sin(2*np.pi*F2*time)
# Noise
#x += noise
```

```python
plt.plot(time, x)
plt.show()
```

```python
x = np.reshape(x, [-1,1])
time = np.reshape(time, [-1,1])
```

```python
data = np.concatenate((time,x), axis=1)

np.savetxt("/home/tbrownex/data/test/sinWave/sin.csv", data, fmt='%1.4f',delimiter=",",header="Period,Value")
```
