#### Creates a very simple time series that should be easily forecast-able

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avg = 100
length = 1000
values = avg + 15*np.random.rand(length)
```

```python
# Put some seasonality in there: every 4 months double the value
frequency = 4
factor = 2
quarters = values[::frequency]
quarters[:] = quarters*factor
```

```python
dates = pd.date_range('1/1/1990', periods=length, freq='MS')
```

```python
plt.plot(dates[:10], values[:10])
plt.ylim(0,300)
```

```python
s = pd.Series(data=values, index=dates, name="value")
s.to_csv("/home/tbrownex/data/test/seasonal.csv", header=True, index_label="date")
```
